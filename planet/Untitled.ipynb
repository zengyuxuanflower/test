```python
import names
import pandas as pd
import numpy as np
import networkx as nx
import random
from tqdm import tqdm

```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#configure

n_names = 500 # number of "resolved" entities
n_max_variations = 5 # maximum number of unresolved entities per resolved entity

n_min_send = 1 # minimum number of entities each entity sends to (each send = multiple but aggregated transactions)
n_max_send = 3 # maximum number of entitise each entity sends to(each send = multiple but aggregated transactions)

near_nodes = 10 # only transact with nodes with ID within +/- this range to create more realistic looking data
```

```python
G = nx.MultiDiGraph() # Directed graphs with self loops and parallel edges
# generate random resolved-entity nodes

# for i in range(n_names):
#     G.add_node(names.get_full_name()) for i in range(n_names)
_ = [G.add_node(i, name=names.get_full_name()) for i in range(n_names)]
# _ = [G.add_node(i, name=names.get_full_name()) for i in range(n_names)]
```

```python
names.get_full_name()
```

```python
# generate edges
for node in tqdm(G.nodes):
    num_edges_to_create = random.randint(n_min_send, n_max_send)
    tmp_entity_list = [] # store beneficiary entities so as not to repeat
    for i in range (num_edges_to_create):
        # pick random entity to transact with - could be self (these are "resolved" entities)
        node_added = False
        while node_added == False:
            # pick random node that is "close" in id to sending node
            # random_node = random.choice(list(G.nodes))
            random_node = random.choice(list([n for n,v in G.nodes(data=True) if node-near_nodes <= n <= node+near_nodes]))
            if random_node not in tmp_entity_list:
                # attributes = {'num_trxn': random.randint(1, 10)}
                trxn_ct = random.randint(1, 100)
                trxn_amt = trxn_ct*random.randint(1, 1000)*1000
                G.add_edge(node, random_node, ntrxn = trxn_ct, vtrxn = trxn_amt, value=trxn_amt)
                tmp_entity_list.append(random_node)
                node_added = True
```

```python
for node1, node2, data in G.edges(data=True):
    if node1 == node2: # check self transaction?
        #print(node1, node2, data['ntrxn'])
        print(node1, G.nodes[node1], '-->', node2, G.nodes[node2], ', n_trxn:', data['ntrxn'], ', value:',data['value'])
```
